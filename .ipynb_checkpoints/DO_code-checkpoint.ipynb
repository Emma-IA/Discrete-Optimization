## Question 1

```python

def columns_q1(dataset, column): #returns a list with, for each products, corresponding interval for the feature "column"
    dataset_column=normalized_df[[str(column),'51']]
    sorted_array=dataset_column.sort_values(by = str(column))
    nb_rows=dataset.shape[0]
    index_list = sorted_array.index
    list_intervals=(-1)*np.ones(nb_rows, dtype=object)
    first=0.0
    last=0.0
    index = 0
    if sorted_array.iloc[0][1]<0.13: #first row
        last= sorted_array.iloc[0,0]
        index=0
    else:
        index=-1
    list_intervals[index_list[0]]=(first, last)
    for i in range(1, nb_rows-1): #intermediate rows
        if sorted_array.iloc[i][1]<0.13:
            if index == i-1: 
                last = sorted_array.iloc[i+1,0]
            else:   
                first = sorted_array.iloc[i-1,0]
                last = sorted_array.iloc[i+1,0]
                
            index=i  
            list_intervals[index_list[i]]=(first+0.000000000000000001, last-0.000000000000000001)
            
    if sorted_array.iloc[nb_rows-1][1]<0.13: #last row
        last= 1
        list_intervals[index_list[nb_rows-1]]=(first, last)

    return list_intervals

        
print("column 12 upper and lower bounds:", columns_q1(normalized_df, 12))
```

```python

def intervals(dataset): #returns a 2D array with intervals (lower bound, upper bound) for each products(rows) and corresponding features(columns)
    nb_rows=dataset.shape[0]
    nb_columns=dataset.shape[1]
    list_intervals=np.zeros((nb_rows, nb_columns-1), dtype=object)
    for j in range(1, nb_columns):
        column_j=columns_q1(dataset, j)
        for i in range(nb_rows):
            if column_j[i] !=-1 : 
                list_intervals[i][j-1]=column_j[i]  
            
    return list_intervals
        

print( "intervals for all columns: ", intervals(normalized_df))

```

```python
 
def sum_max(dataset): # returns the value of the maximum sum expected, and the lower and upper bounds corresponding. 
    shape=dataset.shape
    data=intervals(dataset)
    max_sum=0
    max_row=0
    for i in range(shape[0]):
        sum=0
        for j in range(shape[1]-1):
            if data[i][j]!=0:
                sum+=data[i][j][1]-data[i][j][0]
        if sum>max_sum: 
            max_sum=sum
            max_row=i
    return max_sum, data[max_row]

print("maximum sum:", sum_max(normalized_df)[0])
print("intervals corresponding:", sum_max(normalized_df)[1])
    
```

## Question 3

```python

def columns_q3(array, column): #returns a list with, for each products, corresponding interval's number for the feature "column"
    array_column=array[[str(column),'51']]
    sorted_array=array_column.sort_values(by = str(column))
    nb_rows=array.shape[0]
    index_list = sorted_array.index
    list_subpackages=np.zeros(nb_rows, dtype=object)
    list_nb=(-1)*np.ones(nb_rows, dtype=object) 
    first=0
    last=0
    index = 0
    grid=[]
    for j in range(nb_rows):
        grid.append([])
    counter=0
    if sorted_array.iloc[0][1]<0.9:
        last= sorted_array.iloc[1][0]
        index=0
        grid[counter].append(index_list[0])
        list_nb[index_list[0]]=counter
        
    else:
        index=-1
        counter=-1
    for i in range(1, nb_rows-1):  
        if sorted_array.iloc[i][1]<0.9:
            if index == i-1:
                last = sorted_array.iloc[i+1,0]
                grid[counter].append(index_list[i])
                list_nb[index_list[i]]=counter

            else: 
                first = sorted_array.iloc[i-1,0]
                last = sorted_array.iloc[i+1,0]
                counter+=1
                grid[counter].append(index_list[i])
                list_nb[index_list[i]]=counter
            index=i
            list_subpackages[index_list[i]]=(first, last)
            
     
    if sorted_array.iloc[nb_rows-1][1]<0.9:
        last= 1
        grid[counter].append(index_list[nb_rows-1])
        list_nb[index_list[nb_rows-1]]=counter

    return list_nb

print("column 8 intervals'numbers:", columns_q1(normalized_df, 8))
```

```python

def intervals_numbers(array): #returns a 2D array with intervals's numbers for each products(rows) and corresponding features(columns)
    nb_rows=array.shape[0]
    nb_columns=array.shape[1]
    grid_packages=np.zeros((nb_rows, nb_columns-1), dtype=object)
    
    for j in range(1, nb_columns):
        column_j=columns_q3(array, j)
        for i in range(nb_rows):
            grid_packages[i][j-1]=column_j[i]
                
    return grid_packages
        
print( "intervals'numbers for all columns: ", intervals_numbers(normalized_df))
```

```python

def check_rows(array): # returns the cardinality of the box, and the lower and upper bounds corresponding. 
    grid=intervals_numbers(array)
    nb_rows=array.shape[0]
    row_max=0
    c_max=0
    for i in range(nb_rows-3):#to change
        c_row=0
        for j in range(i+1, nb_rows-2):#to change
            if grid[i][0]==grid[j][0] and grid[i][0]!=-1 :
                if all(grid[i]==grid[j]):
                    c_row+=1
                        
        if c_max<c_row:
            c_max=c_row
            row_max=i
    return c_max+1, row_max

print( "cardinality of the Box:", check_rows(normalized_df)[0] )
print( "upper and lower bounds corresponding: ", intervals(normalized_df)[check_rows(normalized_df)[1]])
```
