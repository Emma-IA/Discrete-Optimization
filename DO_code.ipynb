```python
#for 1 column, if the row corresponds to a good point then we put a list of pairs (li, ui) containing the feature of the good point
#As bounds, I always chose the value of the bad points' features surrounding the packages of good points ... 
#maybe we should round it? Or is it ok to hae a bad point at the border? 

## pbm 5th column !!!!For first columns mainly (I didn't check all of them)

def list_packages_column(array, column):
    array_column=normalized_df[[str(column),'51']]
    sorted_array=array_column.sort_values(by = str(column))
    nb_rows=array.shape[0]
    index_list = sorted_array.index
    list_subpackages=np.zeros(nb_rows, dtype=object)
    first=0
    last=0
    index = 0
    if sorted_array.iloc[0][1]<0.13:
        last= sorted_array.iloc[1][0]
        index=0
    else:
        index=-1
    for i in range(1, nb_rows-1):
        if sorted_array.iloc[i][1]<0.13:
            if index == i-1: 
                last = sorted_array.iloc[i+1,0]
            else:   
                first = sorted_array.iloc[i-1,0]
                last = sorted_array.iloc[i+1,0]
                
            index=i  
                
           # list_subpackages[index_list[i]]=(first+0.000000000000000001, last-0.000000000000000001)
            list_subpackages[index_list[i]]=(first, last)
            
    if sorted_array.iloc[nb_rows-1][1]<0.13:
        last= 1
        list_subpackages[index_list[nb_rows-1]]=(first, last)

    return list_subpackages

        
print(list_packages_column(normalized_df, 1))
```

```python
#gathers all of the columns into 1 2D array with 0 or (li,ui) pairs

def list_packages(array):
    nb_rows=array.shape[0]
    nb_columns=array.shape[1]
    list_packages=np.zeros((nb_rows, nb_columns-1), dtype=object)
    for j in range(1, nb_columns):
        column_j=list_packages_column(array, j)
        for i in range(nb_rows):
            if column_j[i] !=0 : 
                list_packages[i][j-1]=column_j[i]
    
            
    return list_packages
        

list_packages(normalized_df)

```

```python
#takes the array created above, and for each row computes the sum(ui -li) for i from 1 to D ... 
# And finds the sum max + the row corresponding + all the pairs of bounds corresponding. 

def sum_max(array):
    shape=array.shape
    data=list_packages(array)
    max_sum=0
    max_row=0
    for i in range(shape[0]):
        sum=0
        for j in range(shape[1]-1):
            if data[i][j]!=0:
                sum+=data[i][j][1]-data[i][j][0]
        if sum>max_sum: 
            max_sum=sum
            max_row=i
    return max_sum, max_row, data[max_row]

sum_max(normalized_df)
    
```
